```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("https://raw.githubusercontent.com/VictorIvanPizarro/Hito_VisualizacionDatos/main/presidents.csv") #guardar los datos del csv en un DataFrame (estructura de dos dimensiones)
print(df)
df.describe()
```

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plot

df['state'].value_counts().plot(kind="bar", fontsize = 15, color=['red', 'orange', 'blue'], figsize=(20,10))
plot.xticks(rotation = 90)
plot.xlabel('Estados')
plot.ylabel('Candidaturas')
plot.title('Total de candidatos con su Estado')
```

```python
cs = df.groupby('party_detailed')['totalvotes'].sum()
pr = cs.loc[cs > 600000000]
# le digo que solo me muestre los superiores a 600M. Esto debido a que hay muchos y solo nos interesan los de mayor cantidad de votos
pr.sort_values().plot(kind='barh', figsize=(20,10), fontsize=18, color=['green', 'yellow'])
plot.grid(axis='x')
plot.title('Partidos con mayor numero de votos en total', fontsize=18)
plot.ylabel('Partido', fontsize=18)
plot.xlabel('Total de votos en la historia', fontsize=18)
```

```python
cs = df.groupby('party_detailed')['totalvotes'].mean()
pr = cs.loc[cs > 6000000]
pr.sort_values().plot(kind='barh', figsize=(20,10), fontsize=18, color=['blue', 'yellow', 'orange'])
plot.grid(axis='x')
plot.title('Partidos con mayor numero de votos de media', fontsize=18)
plot.ylabel('Partido', fontsize=18)
plot.xlabel('Media de sus votos', fontsize=18)
plot.savefig('customers.png')
```
